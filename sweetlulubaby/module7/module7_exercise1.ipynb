## Data Visualization

Display sample images with bounding boxes.

```python
def plot_image_with_boxes(image, boxes, ax=None):
    """
    Plot an image with bounding boxes
    boxes: numpy array with shape (n, 5) - [class_id, x_center, y_center, width, height]
    All box coordinates are normalized (0-1)
    """
    if ax is None:
        fig, ax = plt.subplots(1, 1, figsize=(8, 8))

    ax.imshow(image)
    height, width = image.shape[:2]

    # Draw each bounding box
    for box in boxes:
        class_id, x_center, y_center, box_width, box_height = box

        # Convert from normalized YOLO format to pixel coordinates
        x_center_px = x_center * width
        y_center_px = y_center * height
        box_width_px = box_width * width
        box_height_px = box_height * height

        # Calculate top-left corner
        x1 = x_center_px - box_width_px / 2
        y1 = y_center_px - box_height_px / 2

        # Draw rectangle
        rect = patches.Rectangle(
            (x1, y1), box_width_px, box_height_px,
            linewidth=2, edgecolor='red', facecolor='none'
        )
        ax.add_patch(rect)

        # Add label
        ax.text(x1, y1-5, 'boat', color='red', fontsize=10,
                bbox=dict(boxstyle='round', facecolor='white', alpha=0.7))

    ax.axis('off')
    return ax
```

```python
# Display 5 sample images with bounding boxes
fig, axes = plt.subplots(2, 3, figsize=(15, 10))
axes = axes.flatten()

for i in range(6):
    plot_image_with_boxes(X[i], y[i], ax=axes[i])
    axes[i].set_title(f'Image {i}: {len(y[i])} boat(s)')

plt.tight_layout()
plt.show()
```

```python
# Fine-tune the YOLO model on the provided (X, y) dataset
# This cell:
# 1) Builds a YOLO-style dataset on disk from arrays X (images) and y (YOLO labels)
# 2) Writes a dataset YAML
# 3) Trains YOLOv8 and saves best weights under runs/detect/boats_yolov8/weights/best.pt

import os
from pathlib import Path
import numpy as np
from PIL import Image

# Reuse the loaded model if available; else load
try:
    model
except NameError:
    from ultralytics import YOLO
    model = YOLO('yolov8n.pt')

# ------------- Build dataset on disk -------------
root = Path('boats_dataset')
for sub in ['images/train', 'images/val', 'labels/train', 'labels/val']:
    (root / sub).mkdir(parents=True, exist_ok=True)

# 80/20 split
idxs = np.arange(len(X))
np.random.seed(42)
np.random.shuffle(idxs)
split = int(0.8 * len(idxs))
train_idx, val_idx = idxs[:split], idxs[split:]

def ensure_rgb_uint8(arr):
    arr = np.array(arr)
    if arr.ndim == 2:
        arr = np.stack([arr]*3, axis=-1)
    if arr.dtype != np.uint8:
        if np.issubdtype(arr.dtype, np.floating):
            # assume in [0,1] or [0,255]
            if arr.max() <= 1.0:
                arr = (np.clip(arr, 0, 1) * 255).astype(np.uint8)
            else:
                arr = np.clip(arr, 0, 255).astype(np.uint8)
        else:
            arr = arr.astype(np.uint8)
    return arr

def save_split(indices, split_name):
    for idx in indices:
        img = ensure_rgb_uint8(X[idx])
        img_path = root / f'images/{split_name}/{idx:05d}.png'
        Image.fromarray(img).save(img_path)
        # y[idx]: array of (n,5) -> class_id, x_center, y_center, width, height (normalized)
        labels = np.array(y[idx])
        label_path = root / f'labels/{split_name}/{idx:05d}.txt'
        with open(label_path, 'w') as f:
            if labels.size > 0:
                for row in labels:
                    cls, xc, yc, w, h = row
                    f.write(f"{int(cls)} {float(xc):.6f} {float(yc):.6f} {float(w):.6f} {float(h):.6f}\n")

save_split(train_idx, 'train')
save_split(val_idx, 'val')

# ------------- Dataset YAML -------------
yaml_path = Path('boats.yaml')
yaml_path.write_text(f"""path: {root.resolve()}
train: images/train
val: images/val
names:
  0: boat
""")

print('Dataset prepared at:', root.resolve())
print('YAML:', yaml_path.read_text())

# ------------- Train -------------
# You can adjust epochs/imgsz/batch according to your machine.
from ultralytics import YOLO
model = YOLO('yolov8n.pt')  # start from COCO-pretrained
results = model.train(
    data=str(yaml_path),
    epochs=30,
    imgsz=640,
    batch=16,
    name='boats_yolov8',
    patience=10,
    save=True,
    verbose=True
)
print('Training finished. Best weights should be under runs/detect/boats_yolov8/weights/best.pt')
```

```python
# Load best model
model = YOLO('runs/detect/boats_yolov8/weights/best.pt')

# Evaluate on validation set
metrics = model.val(data=str(yaml_path))

print('\nValidation Metrics:')
print(f'  mAP50: {metrics.box.map50:.3f}')
print(f'  mAP50-95: {metrics.box.map:.3f}')
print(f'  Precision: {metrics.box.mp:.3f}')
print(f'  Recall: {metrics.box.mr:.3f}')
```
