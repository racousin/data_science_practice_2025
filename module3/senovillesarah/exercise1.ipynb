```python
import pandas as pd
import requests

# URLs
train_url = 'https://www.raphaelcousin.com/modules/data-science-practice/module3/exercise/module3_exercise_train.csv'
test_url  = 'https://www.raphaelcousin.com/modules/data-science-practice/module3/exercise/module3_exercise_test.csv'

def download(url, filename):
    r = requests.get(url)
    r.raise_for_status()
    with open(filename, 'wb') as f:
        f.write(r.content)
    print(f"{filename} t√©l√©charg√©")

download(train_url, 'module3_exercise_train.csv')
download(test_url, 'module3_exercise_test.csv')

# Charger les fichiers
df_train = pd.read_csv('module3_exercise_train.csv', index_col='id')
df_test  = pd.read_csv('module3_exercise_test.csv', index_col='id')
```

```python
df_train = df_train.fillna(-1)
df_test  = df_test.fillna(-1)

X_train = df_train.drop('SalePrice', axis=1)
y_train = df_train['SalePrice']
X_test = df_test.copy()
```

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
import pandas as pd

# üîß Transformation des colonnes non num√©riques
X_train = pd.get_dummies(X_train, drop_first=True)
X_test  = pd.get_dummies(X_test, drop_first=True)

# üîß Alignement pour avoir les m√™mes colonnes dans train et test
X_train, X_test = X_train.align(X_test, join='left', axis=1, fill_value=0)

# üî® Mod√®le de r√©gression lin√©aire
model = LinearRegression()
model.fit(X_train, y_train)

# üìä √âvaluation sur train
y_pred_train = model.predict(X_train)
mae = mean_absolute_error(y_train, y_pred_train)
print("MAE train:", mae)

# üìä Pr√©dictions sur le test set
y_pred_test = model.predict(X_test)

```

```python
submission = pd.DataFrame({
    'id': X_test.index,
    'SalePrice': y_pred_test
})
submission.to_csv('submission.csv', index=False)
submission.head()
```
