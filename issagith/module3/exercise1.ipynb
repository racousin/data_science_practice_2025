```python
import pandas as pd
import numpy as np
```

### Data Collection

```python
df_train =  pd.read_csv("data/module3_exercise_train.csv", sep=",", index_col='id')
```

### Data Preprocessing

```python
def quick_preprocessing(df: pd.DataFrame):
    df = df[df.select_dtypes(include=[np.number]).columns.to_list()]
    df = df.dropna()
    df.fillna(-1)
    return df
```

```python
df = quick_preprocessing(df_train)
```

```python
X_train = df.drop('SalePrice', axis=1)
y_train = df["SalePrice"]
```

```python
X_train.shape, y_train.shape
```

### Model Building and Evaluation

```python
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_absolute_error
```

```python
model = LinearRegression()
model.fit(X_train, y_train)
```

```python
preds = model.predict(X_train)
mean_absolute_error(y_train, preds)
```

```python
print(preds[:5])
print(y_train[:5].to_list())
```

### Generating Submission File

```python
df_X_test = pd.read_csv("data/module3_exercise_test.csv", sep=",", index_col='id')
```

```python
X_test = quick_preprocessing(df_X_test)
```

```python

submission = pd.DataFrame({
    'id': X_test.index,
    'SalePrice': model.predict(X_test)
})

submission.to_csv('submission.csv', index=False, sep=',')
submission.head()
```
